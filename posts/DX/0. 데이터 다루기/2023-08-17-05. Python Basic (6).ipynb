## 파일 쓰기 (`writelines()`)

```python
l = ["이강철\n","28세\n", "통계학 전공\n"]

f = open("test2.txt","w")

f.writelines(l)
f.close()
```

```python
f = open("test2.txt","r")
print(f.read())
```

## 파일 읽기(`readlines()`)

```python
f = open("test2.txt","r")
result = f.readlines() 
print(result)
```

```python
for txt in result:
    print(txt,end="")
```

## 파일 읽기 (`readline() : 한 행씩`)

```python
f = open("test2.txt","r")

result = f.readline()
while result : 
    print(result,end="")
    result = f.readline()
f.close()
```

# 워드 클라우드

```python
txt ="The generative model can be thought of as analogous to a team of counterfeiters, trying to produce fake currency and use it without detection, while the discriminative model is analogous to the police, trying to detect the counterfeit currency. Competition in this game drives both teams to improve their methods until the counterfeits are indistiguishable from the genuine article"
```

```python
txt
```

`-` 공백을 기준으로 단어 생성

```python
word = txt.split(" ")
```

`-` 단어 중복 제거

```python
w_u = set(word)
```

`-` 단어와 개수를 쌍으로 하는 딕셔너리 생성

```python
w_c = {i : word.count(i) for i in w_u}
```

```python
w_c
```

`-` 불용어 삭제

```python
del_word = ["the",'The','a','is','are', 'not','of','on','that','this','and','be','to', 'from']
```

```python
for w in del_word :
    if w in w_c :
        del w_c[w]
```

```python
w_c
```

## 워드클라우드 생성

```python
%config InlineBackend.figure_format='retina' ## 이미지 포맷

# 워드 클라우드 만들기
wordcloud = WordCloud(font_path = 'C:/Windiws/fonts/HMKMRHD.TTF',  ## 글씨 폰트
                      width=2000,
                      height=1000,
                      background_color='white').generate_from_frequencies(w_c) ##우리가 만든 워드 카운트를전달

# 표시하기
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

# 이메일 보내기

## import

```python
import smtplib
from email.mime.text import MIMEText
```

## step1. 기본 셋팅

1. 지메일, 모든 설정 $\to$ 전달 및 POP/IMAP $\to$ IMAP 엑세스(IMAP 사용 체크) $\to$ 변경사항 저장

2. 구글 계정 괸리 클랙 $\to$ 보안 $\to$ 2단계 인증 클릭  $\to$ 앱 비밀번호 $\to$ windows, 컴퓨터 메일 오픈

3. 컴퓨터용 앱 비밀번호 복사

4. 복사 후 첫 번째 인자에는 나의 이메일을, 두 번째 인자에는 복사한 비밀번호를 기입 후 실행

```python
# smtp 주소
s = smtplib.SMTP('smtp.gmail.com', 587)

# TLS 보안 시작
s.starttls() 

## 로그인
s.login('rkdcjf8232@gmail.com', 'zfkdrzytshqsucln') 
```

## step2.  메일 내용구성 

```python
## 본문
m = MIMEText('''
메일 보내기 테스트 테스트
'''
)

## 제목
m["Subject"] = "gc test"
```
