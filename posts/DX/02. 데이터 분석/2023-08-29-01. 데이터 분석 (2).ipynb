## 분산과 표준편차

*  값들이 평균으로부터 얼마나 벗어나 있느지를 나타내는 값.

* 위 값들은 집단간 평균 비교시 사용되는 값들이다.

## 표준오차

$$\text{표준오차} = \frac{s}{\sqrt n},\quad \text{표본분산}=  s^2 = \frac{\sum (x-\bar x )^2}{n-1}$$

`-` [표본의 분산을 $n-1$로 나누는 이유](https://m.blog.naver.com/sw4r/221021838997)

## 95%신뢰구간

```python
rd.normalvariate?
```

`1` 평균이 172, 표준편차가 1인 표본을 800,000개 추출

```python
pop = [round(rd.normalvariate(172, 7),1) for i in range(800000)]
```

```python
sns.histplot(pop, bins = 100)
plt.axvline(np.mean(pop), color = 'r')
plt.text(np.mean(pop)+1, 30000, f'pop mean : {np.mean(pop).round(3)}', color = 'r')
plt.show()
```

`2` . 50개씩 데이터를 추출 후 표본 평균을 계산

```python
x_mean = []
for i in range(100):
    s1 = rd.sample(pop, 50)
    s1 = pd.Series(s1)
    x_mean.append(round(s1.mean(),3))
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(x_mean)
plt.axvline(np.mean(x_mean), color = 'r')
plt.text(np.mean(x_mean)+1, 0.1, f'pop mean : {round(np.mean(x_mean),3)}', color = 'r')
```

## 중심극한 정리

`-` $n$의 크기가 충분히 크면 표본평균의 분포는 정규분포로 수렴한다.

* 조건 1 : 데이터의 수가 충분히 커야함

* 조건 2 :  추출된 표본들은 서로 독립이어야 한다.

### 표본평균의 분포

```python
pop = [round(rd.expovariate(.3)+165,2) for i in range(10001)]
```

```python
# 표본의 크기
n = 100

# 표본의 갯수
m = 200

sample_mean = [np.mean(rd.sample(pop,n)) for i in range(m)]

plt.figure(figsize=(10,6))
sns.kdeplot(sample_mean)

plt.axvline(x=np.mean(sample_mean), color = 'red') #표본평균들의 평균
plt.axvline(x=np.mean(pop), color = 'grey') # 모평균

plt.text(np.mean(sample_mean)-.2, 0.02, round(np.mean(sample_mean),3), color = 'red') #표본평균들의 평균
plt.text(np.mean(pop)+.1,0.02, round(np.mean(pop),3), color = 'grey') #모평균
```

### 모집단의 분포

```python
plt.figure(figsize=(10,6)) #설정
sns.histplot(pop, bins = 100)
plt.axvline(x=np.mean(pop), color = 'grey') # 모평균
plt.text(np.mean(pop)+.5, 800, round(np.mean(pop),2), color = 'grey')
plt.show()
```

## 3. 카이제곱검정

`-` A와 B라는 두 개의 범주형 변수가 존재할 때

* 두 변수들 사이의 관계가 독립인지 아닌지 검정

$$H_0 : \text{두 변수는 독립이다.}\quad H_1 : \text{두 변수는 독립이 아니다.}$$

$$\text {카이제곱 통계량} : \chi^2 [(p-1)(q-1),\alpha]= \sum_{p=1}^P\sum_{q=1}^Q\frac {(\text{관측빈도-기대빈도)}^2}{기대빈도}$$

$$p,q  = \text {각 범주의 클래스의 수}$$

$$n=(p-1)(q-1)\to \text{자유도}$$

$$\text{기대빈도}= \frac{\text{sum}(p=i) \times \text{sum}(q=i)}{\text{total sum}}$$

### 카이제곱통계량 계산(손계산)

`1`. 기대빈도 계산

```python
tem = titanic.groupby(["Pclass","Survived"],as_index=False)[["Pclass"]].value_counts()
tem.rename(columns = {"count": "관측빈도"},inplace=True)
```

```python
tem
```

```python
tem1  = tem.groupby(["Pclass"],as_index=False)[["관측빈도"]].sum().rename(columns ={"관측빈도" : "P_total"})
tem2  = tem.groupby(["Survived"],as_index=False)[["관측빈도"]].sum().rename(columns ={"관측빈도" : "S_total"})
```

```python
total = pd.merge(pd.merge(tem,tem1),tem2)
```

```python
total["total"] =sum(total.S_total.unique())
```

```python
total["기대빈도"] = total.P_total*total.S_total/total.total
```

```python
total
```

`2`. 카이제곱 통계량 계산

```python
pd.crosstab(titanic['Survived'], titanic['Pclass'])
```

```python
chi_s = sum(((total["관측빈도"] - total["기대빈도"])**2)/total["기대빈도"])
chi_s
```

```python
alpha = 0.05 ## 유의수준
```

```python
a2 = spst.chi2.ppf(1-alpha/2,2) ## 기각영역1
a1 = spst.chi2.ppf(alpha/2,2) ## 기각영역 2

(a1,a2)
```

```python
x = np.linspace(0,110,10000) ## 임의의 x생성
s = spst.chi2.pdf(x,2) ## x에 따른 카이제곱 분포 생성
```

```python
#| code-fold : true
plt.plot(x,s,".r",alpha=0.2)
plt.plot(x[(x>a1) & (x<a2)],s[(x>a1) & (x<a2)],".b",alpha=0.3)
plt.axvline(x=chi_s,color="b",linestyle = ":")
plt.text(chi_s-15,.4, r"계산된 $\chi$ 의 위치 "+" \n = "+str(round(chi_s,2)))
plt.legend(["빨강(기각영역)","파랑(신뢰구간)"])
```
