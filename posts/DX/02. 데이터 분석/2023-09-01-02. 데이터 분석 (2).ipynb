# 데이터 로드

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/titanic.1.csv')
```

## 시각화 1. (`Survived <-> Age`)

1. `common_norm = True (기본값)` : 칵 클래스의 비율을 합쳐서 표시 (두 그래프의 아래 면적의 합이 1)

2. `common_norm = False`: 각각의 클래스로 니누어서 비율을 표시 (각 클래스의 아래 면적은 1)

3. `multiple = ‘fill’`: 각 클래스에 따른 비율을 비교 (각 클래스의 아래 면적은 1)

```python
titanic.Survived.value_counts(normalize=True)
```

```python
#| code-fold : true
fig, axes = plt.subplots(2,3,figsize=(12,8))

(ax1,ax2,ax3), (ax4,ax5,ax6) = axes

sns.histplot(x='Age', data = titanic, hue = 'Survived',ax=ax1,stat="density")
sns.histplot(x='Age', data = titanic, hue = 'Survived', common_norm = False,ax=ax2,stat="density")
sns.histplot(x='Age', data = titanic, hue = 'Survived', multiple="fill",ax=ax3)

ax3.axhline(titanic['Survived'].mean(), color = 'r')
sns.kdeplot(x='Age', data = titanic, hue ='Survived',ax=ax4)
sns.kdeplot(x='Age', data = titanic, hue ='Survived',
            common_norm = False,ax=ax5)
sns.kdeplot(x='Age', data = titanic, hue ='Survived'
            , multiple = 'fill',ax=ax6)
ax6.axhline(titanic['Survived'].mean(), color = 'r')
fig.tight_layout()
```

## 시각화 2. (`Survived <-> Fare`)

```python
# | code-fold : true
fig, axes = plt.subplots(2,3,figsize=(12,8))

(ax1,ax2,ax3), (ax4,ax5,ax6) = axes

sns.histplot(x='Fare', data = titanic, hue = 'Survived',bins=40,ax=ax1,stat="density")
sns.histplot(x='Fare', data = titanic, hue = 'Survived',bins=40, common_norm = False,ax=ax2,stat="density")
sns.histplot(x='Fare', data = titanic, hue = 'Survived', multiple="fill",ax=ax3,bins=40)

ax3.axhline(titanic['Survived'].mean(), color = 'r')
sns.kdeplot(x='Fare', data = titanic, hue ='Survived',ax=ax4)
sns.kdeplot(x='Fare', data = titanic, hue ='Survived',
            common_norm = False,ax=ax5)
sns.kdeplot(x='Fare', data = titanic, hue ='Survived'
            , multiple = 'fill',ax=ax6)
ax6.axhline(titanic['Survived'].mean(), color = 'r')
fig.tight_layout()
```
